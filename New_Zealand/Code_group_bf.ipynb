```python
#Auckland Active life pull
def active_life(set_num): 

    url = 'https://api.yelp.com/v3/businesses/search'
    headers = {
        'Authorization': f"Bearer {naveenkey}",}
    url_params = {
    "categories" : "active, All",
    "location" : "Auckland",
    "offset" : offset_num,
    "limit" : 50}

    response = requests.get(url, headers = headers, params = url_params)
    return response.json()

for offset_num in np.arange(50,400,50) : 
    try:
        output_json = active_life(offset_num)
        #pprint(offset_num) 
        #pprint(output_json)
        if offset_num == 50:
            df1a = pd.DataFrame.from_dict(output_json['businesses'])
        else:
            df2a = pd.DataFrame.from_dict(output_json['businesses'])
            df1a = df1a.append(df2a)
    except AttributeError:
        pprint("error at ", offset_num) 
```

```python
#Auckland_restaurants
def Auckland_restaurants(set_num):
    url = 'https://api.yelp.com/v3/businesses/search'
    headers = {
        'Authorization': f"Bearer {yelp_api}",
    }
    url_params = {
        "categories" : "restaurants, All",
        "location" : "Auckland",
        "offset" : offset_num,
        "limit" : 50}
    
    response_auck_rest = requests.get(url, headers = headers, params = url_params)
    return response_auck_rest.json()


for offset_num in np.arange(50,1000,50) :
    try:
        output_json = Auckland_restaurants(offset_num)
        #pprint(offset_num)
        #pprint(output_json)
        if offset_num == 50:
            df_Auckland_restaurants = pd.DataFrame.from_dict(output_json['businesses'])
        else:
            df2 = pd.DataFrame.from_dict(output_json['businesses'])
            df_Auckland_restaurants = df_Auckland_restaurants.append(df2)
    except AttributeError:
           print("error at ", offset_num)
```

```python
NZ = df_New_Zealand["review_count"].sum()

#Acukland
NZ_Auckland = df_New_Zealand.loc[(df_New_Zealand["city"] == 'Auckland')]
NZ_Auckland_group = NZ_Auckland.groupby("rating")
NZ_Auckland_count = NZ_Auckland_group.review_count.count()
NZ_Auckland_sum = NZ_Auckland_group.rating.sum()
NZ_Auckland_avg_rating = (NZ_Auckland_sum/NZ_Auckland_count)
#Wellington
NZ_Wellington = df_New_Zealand.loc[(df_New_Zealand["city"] == 'Wellington')]
NZ_Wellington_group = NZ_Wellington.groupby("rating")
NZ_Wellington_count = NZ_Wellington_group.review_count.count()
NZ_Wellington_sum = NZ_Wellington_group.rating.sum()
NZ_Wellington_avg_rating = (NZ_Wellington_sum/NZ_Wellington_count)
#Christchurch
NZ_Christchurch = df_New_Zealand.loc[(df_New_Zealand["city"] == 'Christchurch')]
NZ_Christchurch_group = NZ_Christchurch.groupby("rating")
NZ_Christchurch_count = NZ_Christchurch_group.review_count.count()
NZ_Christchurch_sum = NZ_Christchurch_group.rating.sum()
NZ_Christchurch_avg_rating = (NZ_Christchurch_sum/NZ_Christchurch_count)

# Build the scatter plots for each city
Auckland_city=plt.scatter(NZ_Auckland_count, NZ_Auckland_avg_rating, marker="x", facecolors="#FF7F50", edgecolors="black",
            alpha=0.7, label='Auckland')

Wellington_city=plt.scatter(NZ_Wellington_count, NZ_Wellington_avg_rating, marker="o", facecolors="#ffd700", edgecolors="black",
           alpha=0.7,  label='Wellington')

Christchurch_city=plt.scatter(NZ_Christchurch_count, NZ_Christchurch_avg_rating, marker="^", facecolors="#87CEEB", edgecolors="black",
          alpha=0.7, label='Christchurch')
# Incorporate the other graph properties
plt.grid(b=None, which='major', axis='both',linewidth=1)
plt.title("New Zealand Ratings")
plt.xlabel("Total Number of Ratings")
plt.ylabel("Rating")
# Create a legend
plt.legend(handles=[Auckland_city, Wellington_city,Christchurch_city], title= "City's", loc="best")
# Incorporate a text label regarding circle size
#textstr = 'Note:\nCircle size correlates with driver count per city\n'
#plt.text(41.2, 33.5, textstr, fontsize=10)

# Save Figure
#plt.savefig("../Images/Ride_Sharing_Data.png")
# Show Figure
plt.show()
```
