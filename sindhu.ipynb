```python
import requests
import json
from pprint import pprint
import numpy as np
import pandas as pd

from sconfig import sindhu_api


url = 'https://api.yelp.com/v3/businesses/search'
headers = {'Authorization': f"Bearer {sindhu_api}",}
url_params_wel = {"term" : "shopping centers","location" : "Auckland"}

response_auck = requests.get(url, headers=headers, params=url_params_wel)
auck_data = response_auck.json()
pprint(auck_data)


```

```python
df_shop_auck = pd.DataFrame(auck_data['businesses'])
df_shop_auck.head()
```
