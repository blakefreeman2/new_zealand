```python
import requests
import json
from pprint import pprint
import numpy as np
import pandas as pd

from sconfig import sindhu_api

```

```python
def shop_life(set_num): 

    url = 'https://api.yelp.com/v3/businesses/search'
    headers = {
        'Authorization': f"Bearer {sindhu_api}",}
    url_params = {
    "term" : "shopping centers",
    "location" : "Auckland",
    "offset" : offset_num,
    "limit" : 50}

    response = requests.get(url, headers = headers, params = url_params)
    return response.json()

for offset_num in np.arange(0,200,50) : 
    try:
        output_json = shop_life(offset_num)
        pprint(offset_num) 
        pprint(output_json)
        if offset_num == 50:
            df_shop_auck = pd.DataFrame.from_dict(output_json['businesses'])
        else:
            auck_data = pd.DataFrame.from_dict(output_json['businesses'])
            df_shop_auck = df_shop_auck.append(auck_data)
    except AttributeError:
        pprint("error at ", offset_num)
        
```
